```python
import numpy
from matplotlib import pyplot
```

```python
uniformArray = numpy.random.uniform(5, 50, 1000)
print(uniformArray)
```

```python
pyplot.hist(uniformArray, 100)
pyplot.show()
```

```python
normalArray = numpy.random.normal(100, 0.5, 1000)
print(normalArray)
```

```python
pyplot.hist(normalArray, 100)
pyplot.show()
```
